## 2. Foundations walk

Load the per-paper notation glossary and print a quick summary. Foundation concepts live in [`pleyva2004/math-foundations`](https://github.com/pleyva2004/math-foundations); this cell only prints what the paper actually uses.

```python
import json, pathlib

notation_path = pathlib.Path('notation.json')
if notation_path.exists():
    notation = json.loads(notation_path.read_text())
    print(f"Notation entries used by this paper: {len(notation)}")
    for sym in list(notation)[:10]:
        entry = notation[sym]
        read = entry.get('read_as', '?') if isinstance(entry, dict) else '?'
        meaning = entry.get('meaning', '?') if isinstance(entry, dict) else '?'
        print(f"  {sym:<8} read as {read!r:<24} {meaning}")
else:
    print('No notation.json found — run the notation extractor first.')
```
